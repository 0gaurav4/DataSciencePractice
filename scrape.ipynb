```python
url = f'https://twitter.com/search?q={query}&src=typed_query&f=live'
response = requests.get(url)
html = response.text
```

```python
soup = BeautifulSoup(html, 'html.parser')
tweets = soup.find_all('div', {'class': 'tweet'})

for tweet in tweets[:num_tweets]:
    # extract tweet text
    text = tweet.find('div', {'class': 'tweet-text'}).get_text()
    # extract tweet timestamp
    timestamp = tweet.find('span', {'class': '_timestamp'}).get_text()
    # extract user information
    user_name = tweet.find('span', {'class': 'username'}).get_text()
    user_location = tweet.find('span', {'class': 'tweet-geo-text'})
    if user_location is not None:
        user_location = user_location.get_text()
    # do something with the extracted information
    print(f"{user_name} from {user_location} tweeted: {text} at {timestamp}")
```

### from dputils.scrape import Scraper, Tag

```python
# from dputils.scrape import Scraper, Tag
```

```python
def get_data(query, page=1):
    scraper = Scraper(f'https://www.amazon.in/s?k={query}&page={page}')
    print(scraper.soup.title)
    target= Tag(cls='s-main-slot s-result-list s-search-results sg-row')
    items= Tag(cls='s-result-item')
    brand = Tag('h5')
    title = Tag('h2')
    price = Tag('span',cls='a-price-whole')
    rating = Tag('span', cls='a-size-base')
    oprice = Tag('span', cls='a-offscreen')
    return scraper.get_all(target, items,
        brand=brand, title=title, price=price, 
        rating=rating, oprice=oprice)
```

```python
query = 'watch'

all_data = []
for i in range(1,11):
    print(f'fetching page {i} data')
    data = get_data(query, i)
    if len(data) == 0:
        print('No more data')
        break
    all_data.extend(data)
all_data
    
```

```python
import pandas as pd
pd.DataFrame(all_data).dropna(how='any').to_excel('amazon.xlsx', index=False)
```

```python
query = 'mobiles'
url=f'https://www.amazon.in/s?k={query}'
print(url)
soup = sc.get_webpage_data(url)
```

```python
# target
target = {'tag':'div', 'attrs':{ 'class': 's-main-slot s-result-list s-search-results sg-row' }}
# items
items = {'tag':'div', 'attrs':{ 'class': 's-result-item' }}
# heading
title = {'tag':'h2','attrs':{'class':'a-size-mini a-spacing-none a-color-base s-line-clamp-2'}}
# price
price = {'tag':'span','attrs':{'class':'a-price-whole'}}

rating = {'tag':'div','attrs':{'class':'a-section a-spacing-none a-spacing-top-micro'}}
```

#### chatgpt solution

```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
hashtag = '#python'
num_tweets = 100
```

```python
url = f'https://twitter.com/search?q=%23{hashtag}&src=typed_query&f=live'
response = requests.get(url)
html = response.text
```

```python
soup = BeautifulSoup(html, 'html.parser')
tweets = soup.find_all('div', {'class': 'tweet'})

for tweet in tweets[:num_tweets]:
    # extract tweet text
    text = tweet.find('div', {'class': 'tweet-text'}).get_text()
    # extract tweet timestamp
    timestamp = tweet.find('span', {'class': '_timestamp'}).get_text()
    # extract user information
    user_name = tweet.find('span', {'class': 'username'}).get_text()
    user_location = tweet.find('span', {'class': 'tweet-geo-text'})
    if user_location is not None:
        user_location = user_location.get_text()
    # extract hashtags
    hashtags = re.findall(r'#\w+', text)
    # do something with the extracted information
    print(f"{user_name} from {user_location} tweeted: {text} at {timestamp}")
    print(f"Hashtags: {', '.join(hashtags)}")
```
